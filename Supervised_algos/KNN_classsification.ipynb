### KNN Classification algorithm
using the popular Iris dataset

```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.neighbors import KNeighborsClassifier 
```

```python
iris = datasets.load_iris()
iris.data
```

```python
X = iris.data
y = iris.target
#print(y, X)
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 3)
```

```python
from sklearn import metrics
k_range = range(1, 26)
scores = {}
scores_list = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores[k] = metrics.accuracy_score(y_test, y_pred)
    scores_list.append(metrics.accuracy_score(y_test, y_pred))
```

```python
plt.plot(k_range, scores_list)
plt.xlabel('Values of K for KNN')
plt.ylabel('Testing accuracy')
```

```python
# based on above graph I think the model works good with K as 8, now I'll train the model on the full dataset.
knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(X, y)
```

```python
classes = {0:'Setosa', 1:'Versicolor', 2:'Virgina'}

x_new = [[3, 4, 5, 2],  # dummy data for versicolor
         [5, 4, 2, 2]]  # dummy data for a setosa maybe.

y_pred = knn.predict(x_new)

print(classes[y_pred[0]])
print(classes[y_pred[1]])
```
